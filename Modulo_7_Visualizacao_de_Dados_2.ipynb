# **Visualizando Relações Estatísticas nos Dados.**

> A análise estatística é o processo de entender como variáveis no conjunto de dadosse relacionam uma com a outra e como essas relações dependem de outras variáveis. A Visualização de dados se orna uma peça essencial nesse processo, pois, quando os dados são visualizados adequadramente, as pessoas conseguem descobrir padrões que indicam relações nos dados.



**Função relplot()**

Método de alto nível para visualização de relações estatísticas entre variáveis. O método relplot() já combina o **scatter plot** e **line plots**.

Muito útil para ver relações entre duas ou mais variáveis com o uso de parâmetros d semântica dos dados.

# **Relação de variáveis com scatter plots**

**Visualizando a relação entre a varável valor total da conta e gorjeta**

```python
tips = sns.load_dataset('tips')
sns.relplot(x="total_bill", y ="tip", data = tips);
```

**Visualizando a relação entre as variáveis total da conta e gorjeta e colorindo cada ponto com a variável smoker(fumante)**

```python
sns.relplot(x="total_bill", y = "tip", hue="smoker", data = tips);
```

**Visualizando a relação entre as variáveis total da conta e gorjeta e colorindo cada ponto com a variável smoker(fumante)**

```python
sns.relplot(x="total_bill", y = "tip", hue="sex", data = tips);
```

**Visualizando a relação entre as variáveis total da conta e gorjeta e colorindo cada ponto com a variável sexo e definindo um estilo também vaseado na variável sexo**

```python
sns.relplot(x="total_bill", y = "tip", hue="sex", style="sex", data = tips);
```

**É possível representar quatro variáveis alterando a tnalidade e estilo de cada ponto. Mas é preciso cautela, pois, a visualização pode ficar confusa**

```python
sns.relplot(x="total_bill", y = "tip", hue="sex", style="smoker", data = tips);
```

**Em casos de variáveis numéricas para definir a tonalidade a paleta de cores altera de forma sequencial**

```python
sns.relplot(x="total_bill", y = "tip", hue="size", data = tips);
```

**É possível customiar a paleta de cores sequenciais no link:**
https://seaborn.pydata.org/generated/seaborn.color_palette.html

**Usando a opção size para definir o tamanho de cada ponto baseado em alguma coluna da base de dados**

```python
sns.relplot(x="total_bill", y = "tip", size="size", data = tips);
```

```python
sns.relplot(x="total_bill", y = "tip", size="size", sizes=(15,300), data = tips);
```

# Scatterplot

*   Caso não queira usar o relplot() o método scatterplot() já atende perfeitamente quando queremos ver relações nos dados.
*   A únida diferença é que o scatterplot plota gráfico de dispersão apenas, logo este não contém função lineplot()



**Visualiando relação enre as 4 variáveis total_bill, tipo, day e smoker**

```python
sns.scatterplot(x="total_bill", y="tip", hue="day", style="time", data=tips)
```

# Cores no Seaborn

*   A função que controla as cores no seaborn é a color_palette()
*   Essa função fornece uma interface para gerar cores no seaborn
* Usamos esse método internamente ao usar o parâmetro palette
* O métodolo color_palette() recebe o nome de uma paleta de cores do seaborn ou um colormap do matplotlib
* O método set_palette() recebe os mesmos parâmetros e é usado para definir um estilo padrão para todos os plots.
* As cores dvem ser definidas levando em consideração o tipo de dado que queremos plotar. (qualitativos, sequencianis ou divergentes)

**Ferramenta Color Brewer 2.0**

https://colorbrewer2.org/#type=sequential&scheme=BuGn&n=3

**Paleta de cores para dados quantitativos**

```python
sns.palplot(sns.color_palette("Accent"))
```

**Paleta de cores para dados divergentes**

```python
sns.palplot(sns.color_palette("PuOr"))
```

**Paleta de cores para dados sequenciais**

```python
sns.palplot(sns.color_palette("Blues"))
```

**Gráfico usando a paleta d cor sequencial "Blues"**

```python
sns.relplot(x="total_bill", y="tip", hue="size", palette="Blues", size="size", sizes=(12,250), data=tips)
```

**Gráfico usando a paleta d cor sequencial "Set2"**

```python
sns.relplot(x="total_bill", y="tip", hue="size", palette="Greens", size="size", sizes=(12,250), data=tips)
```

**Gráfico usando a paleta de cor qualitativa "Set2"**

* Paleta de cor ideal para plotar dados categóricos  



```python
sns.catplot(x='sex', kind='count', palette='Set2', data=tips)
```

**Gráfico usando a paleta de cor divergente "PiYG"**

```python
sns.heatmap(tips.corr(), annot=True, cmap="PiYG")
```

**Definindo cores padrão para gráficos Seaborn"**

```python
sns.set_palette('pastel')

#plot usando a paleta de cor "pastel"
sns.relplot(x='total_bill', y='tip', data=tips)
```

**Instrução with para definir cor temporariamente**

```python
#plota o gráfico definindo uma opação de cor temporária
with sns.color_palette('Accent'):
  sns.relplot(x='total_bill', y='tip', size='size', sizes=(12,250), data=tips)
```
